```python
def bar_scatter(df, x, y1, y2, t, city):

    fig=go.Figure()

    pop=go.Scatter(
        x=df[x], y=df[y1],
        mode='lines+markers',
        xaxis = 'x',
        yaxis='y1',
        line_color='rgb(128, 0, 128)',
        name='인구 수 대비 '+t
    )
    fac=go.Bar(
        x=df[x], y=df[y2],
        xaxis = 'x',
        yaxis='y2',
        marker=dict(
            color='rgba(50, 171, 96, 0.6)',
            line=dict(
                color='rgba(50, 171, 96, 1.0)',
                width=1),
        ),
        name=t
    )

    data=[pop, fac]

    layout=dict(

        geo=dict(
            showframe=True,
            showcoastlines=True
        ),

        yaxis=dict(
            side='right',
            overlaying='y2',
            anchor='x',
            rangemode='tozero'
        ),

        yaxis2=dict(
            side='left',
            anchor='x',
            rangemode='tozero'
        ),

        plot_bgcolor='rgb(248, 248, 255)',
        legend=dict(x=0, y=1.15, font_size=11)
    )

    fig=go.Figure(data=data, layout=layout)
    fig.update_layout(title_text=city, title_x=0.5, font=dict(family="Noto Sans KR",size=14))
    return fig
```
